```python
import os 
import glob
```

```python
path = './Dataset/'
i = 1
map_days_files = dict()
#   or file.find("magnetic_field.data.csv.gz")!=-1 or file.find("orientation.data.csv.gz")!=-1 or file.find("gyroscope.data.csv.gz")!=-1 or file.find("gravity.data.csv.gz")!=-1 or file.find("linear_acceleration.data.csv.gz")!=-1 or file.find("rotation_vector.data.csv.gz")!=-1):
for root, dirs, files in os.walk(path):
     for file in files:
        if(file.find("accelerometer.data.csv.gz")!=-1 or file.find("gyroscope.data.csv.gz")!=-1 ):
            if root not in map_days_files:
                map_days_files[root] = list();
            map_days_files[root].append(os.path.join(root,file))
            #print(os.path.join(root,file))
            #print(root)

#print(len(map_days_files))
map_new = dict()
for file_list in map_days_files:
    _list = map_days_files[file_list]
    for l in _list:
        cmd = 'gunzip -k '+l;
        print(cmd)
        os.system(cmd)
        cmd = 'cp '+ l[:-3]+" " +os.path.join('./NewFiles/',"_".join(l.split('/')[2:]))[:-3]
        if file_list not in map_new:
            map_new[file_list] = list();
        map_new[file_list].append(os.path.join('./NewFiles/',"_".join(l.split('/')[2:]))[:-3])
        os.system(cmd)
```

```python
import pandas as pd
import numpy as np
def combine(_list,basename):
    data_list = list()
    for _file in _list:
        if 'rotation' not in _file:
            df = pd.read_csv(_file, header=None,names = ['time','x','y','z','_w','category'])
        else:
            df = pd.read_csv(_file, header=None,names = ['time','x','y','z','_w','_q','_a','category'])
        data_list.append(df)
        print(_file)
    
    index_list = np.zeros((len(_list),), dtype=int)
    max_list = np.zeros((len(_list),),dtype = int)
    for i in range(0,len(_list)):
        max_list[i] = data_list[i].count()['x']
    print(max_list)
    #assumption first is accelerometer
    ans_list = list()
    for index, row in data_list[0].iterrows():
        temp_ans = list()
        temp_ans.append(row['time'])
        temp_ans.append(row['x'])
        temp_ans.append(row['y'])
        temp_ans.append(row['z'])
        
        for i in range(1,len(data_list)):
            #print(data_list[i].count()['x'])
            #print(i)
            if max_list[i] == 0:
                temp_ans.append(0)
                temp_ans.append(0)
                temp_ans.append(0)
                continue;
            _temp = data_list[i].at[index_list[i],'time']
            _temp2 = _temp
            
            flag = False
            if index_list[i]+1 < max_list[i]:
                _temp2 = data_list[i].at[index_list[i]+1,'time'] # need to check for bound
                flag = True
            
            if abs(row['time']-_temp) <= abs(row['time']-_temp2):
                
                temp_ans.append( data_list[i].at[index_list[i],'x'])
                temp_ans.append( data_list[i].at[index_list[i],'y'])
                temp_ans.append( data_list[i].at[index_list[i],'z'])
            else:
                if flag==False:
                    temp_ans.append( data_list[i].at[index_list[i],'x'])
                    temp_ans.append( data_list[i].at[index_list[i],'y'])
                    temp_ans.append( data_list[i].at[index_list[i],'z'])
                else:
                    temp_ans.append( data_list[i].at[index_list[i]+1,'x'])
                    temp_ans.append( data_list[i].at[index_list[i]+1,'y'])
                    temp_ans.append( data_list[i].at[index_list[i]+1,'z'])
                    index_list[i]+=1
                    
        temp_ans.append(row['category'])
        ans_list.append(temp_ans)
    
    df = pd.DataFrame(np.array(ans_list))
    # print(df.head())
    return df
    
```

```python
j = 1
for m in map_new:
    a = 0
    for i in range(0,len(map_new[m])):
        if 'accelerometer' in map_new[m][i]:
            a = i;
            break
    temp = map_new[m][0]
    map_new[m][0] = map_new[m][a]
    map_new[m][a] = temp
    _list = map_new[m];
    
    print("Combining ", j," of ",len(map_new))
    df = combine(_list,m);
    df.to_csv("_".join(m.split("/"))[1:]+".csv", sep=',',index=False)
    j+=1
```

```python
import glob,os
import pandas as pd
path ='./TotalCombined/Train/' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
temp =0 
for file_ in allFiles:
    temp+=1
    df = pd.read_csv(file_,index_col=None)
    list_.append(df)
    print(file_)
frame = pd.concat(list_)
```

```python
out = 'Train'
isAcc = 2
# frame.to_csv(out+'CombinedFile.csv', sep=',')
index = 4
val_check = 0
if (isAcc==0):
    gb = frame.groupby('7')
    index = '7'
elif isAcc==1:
    gb = frame.groupby('4')
    index = '4'
else:
    gb = frame.groupby('22')
    index = '22'
    
for x in gb.groups:
    temp_df = gb.get_group(x)
    print(temp_df.at[0,index])
    if temp_df.at[0,index][0]=='sitting' or temp_df.at[0,index][0]=='s' :
        val_check+=temp_df.shape[0]
        temp_df.to_csv(out+'CombinedFile_sitting.csv', sep=',')
    if temp_df.at[0,index][0]=='laying_down'or temp_df.at[0,index][0]=='l':
        val_check+=temp_df.shape[0]
        temp_df.to_csv(out+'CombinedFile_laying.csv', sep=',')
    if temp_df.at[0,index][0]=='standing':
        val_check+=temp_df.shape[0]
        temp_df.to_csv(out+'CombinedFile_standing.csv', sep=',')
    if temp_df.at[0,index][0]=='walking':
        val_check+=temp_df.shape[0]
        temp_df.to_csv(out+'CombinedFile_walking.csv', sep=',')
```
